# Cleaning

```python
attacks.columns
```

```python
# Erasing the columns that I don't expect to use
attacks_clean = attacks[['Date', 'Year', 'Type', 'Country', 'Area', 'Location',
       'Activity', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Time','Species ']]
```

```python
# Erasing the rows where all columns are NaN
attacks_clean.dropna(how='all',inplace=True)
# df = df[df['EPS'].notna()]
```

```python
attacks_clean # that's better
```

## Now to modify the dates

```python
# I'm not using dates on this project but figured it was worth the practice.
```

```python
novas_colunas = [re.sub('[\. ]','_',colunas.lower().strip()) for colunas in attacks_clean]
```

```python
novas_colunas
```

```python
attacks_clean.columns = novas_colunas
```

```python
attacks_clean['new_date'] = attacks_clean['date'].str.extract(r'(\d{2}-[A-Za-z]{3}-{1}\d{4})') # 03-May-2014
```

```python
attacks_clean.head(50)
```

```python
# new_date_df.loc[new_date_df['date'].str.find('--')!=-1,'date_in_datetime'] = pd.to_datetime(new_date_df['date'])
# new_date_df.loc[~(new_date_df['date'].str.find('--')!=-1),'date_in_datetime'] = pd.to_datetime(new_date_df['date'])
# new_date_df
```

```python
attacks_clean['datetime'] = pd.to_datetime(attacks_clean['new_date'])
```
